```python
import pandas as pd
import numpy as np
```

```python
anime_data=pd.read_csv('anime.csv')
rating_data=pd.read_csv('rating.csv')
rating_data=rating_data.rename(columns={'rating':'user_rating'})
```

```python
anime_data = anime_data.merge(rating_data,on='anime_id')
anime_data = anime_data.rename(columns={'name': 'anime_title'})
```

```python
anime_data.head()
```

```python
anime_data["user_rating"].replace({-1:np.nan}, inplace=True)
```

```python
anime_data.isnull().sum()
```

```python
anime_data = anime_data.dropna(axis = 0, how ='any') 
```

```python
anime_data.isnull().sum()
```

```python
counts = anime_data['user_id'].value_counts()
anime_data = anime_data[anime_data['user_id'].isin(counts[counts >= 100].index)]
```

```python
anime_pivot=anime_data.pivot_table(index='anime_title',columns='user_id',values='user_rating').fillna(0)
anime_pivot.head(10)
```

```python
from scipy.sparse import csr_matrix
```

```python
anime_matrix = csr_matrix(anime_pivot.values)
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(anime_matrix)
```

```python
query_index = np.random.choice(anime_pivot.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(anime_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(anime_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, anime_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
import re
def text_cleaning(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    
    return text

anime_data['anime_title'] = anime_data['anime_title'].apply(text_cleaning)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
anim=pd.read_csv('anime.csv')
```

```python
anim['genre'] = anim['genre'].fillna('')
genres_str = anim['genre'].str.split(',').astype(str)
tfv_matrix = tfv.fit_transform(genres_str)
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
indices = pd.Series(anim.index, index=anim['name']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    anime_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return pd.DataFrame({'Anime name': anim['name'].iloc[anime_indices].values,
                                 'Rating': anim['rating'].iloc[anime_indices].values})
```
