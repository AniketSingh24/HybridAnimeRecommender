```python
import pandas as pd
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
cardb=pd.read_csv("CarPrice_Assignment.csv")
cardb.columns
```

```python
cardbint=cardb[['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','price']]
```

```python
cardb['CarName']=cardb['CarName'].str.replace(' ','')
X=cardb[['CarName','carbody','drivewheel','fueltype','enginetype','horsepower','fuelsystem','citympg','highwaympg','enginesize','boreratio']]
y=cardb.price
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
ohe=OneHotEncoder()
ohe.fit(X[['CarName','drivewheel','fueltype','enginetype','carbody','fuelsystem']])
col= make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['CarName','drivewheel','fueltype','enginetype','carbody','fuelsystem']), remainder='passthrough')
```

```python
model=LinearRegression()
pipe=make_pipeline(col,model)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
